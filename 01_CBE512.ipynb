## **(b)** Processing strings into feature vectors.
Write a function that converts the SMILES string to a numerical "Bag-of-Elements" vector.  Note that you will have to be a little careful with handling 'Br' vs. 'B' and 'Cl' vs. 'C'

```python
# Load the data from GitHub
url = 'https://raw.githubusercontent.com/webbtheosim/CBE512-MLinChmSciEng/main/data/Tm_200_subset.csv'
data = pd.read_csv(url, sep=",")
data.head(10)
```

```python
import re # for regex

# func to convert SMILES string to BoE vector
def smiles_to_bag_of_elements(smiles):
    # element map (excluding hydrogen)
    element_map = {'C': 0, 'O': 1, 'N': 2, 'Cl': 3, 'Br': 4, 'B': 5, 'I': 6, 'F': 7, 'S': 8, 'P': 9}
    
    # dictionary to store element counts
    element_counts = {i: 0 for i in range(10)}
    
    # use regex to find all the elements
    element_pattern = re.compile(r'Br|Cl|[B-IK-NO-Z]')
    
    # find all elements in the SMILES string
    elements = element_pattern.findall(smiles)
    
    # count occurrences of each element
    for elem in elements:
        if elem in element_map:
            index = element_map[elem]
            element_counts[index] += 1
    
    # convert the dictionary to a list and return it as the BoE vector
    return [count for count in element_counts.values()]
```

```python
# now run the function to the 'smiles' column of the data
data['bag_of_elements'] = data['smiles'].apply(smiles_to_bag_of_elements)

# print first five rows to see if it's correct
print(data[['smiles', 'bag_of_elements']].head())
```

## **(c)** Chemical similarity as a norm.

Use the norm induced by the dot product to obtain a  distance as a metric for similarity / dissimilarity. Identify which set of molecules are "most similar" and which set of molecules are "most dissimilar" based on this distance metric. What limitations might there be to using this expression for similarity between molecules?

**Answer:** Some limitations might be that 1) there's no consideration of molecular structure (because of the BoE vector counting the atoms but ignoring the bonding/3D structure and the Euclidean distance only capturing the differences in the number of atoms); 2) overemphasis on atom counts (more weight on differences in the number of atoms than on the specific types of atoms, so even if two molecules are really similar besides number of atoms, it could be very far apart), 3) no handling for functional groups at all either, which probably isn't ideal.

```python
from scipy.spatial.distance import euclidean

most_similar_pair = None
most_dissimilar_pair = None
min_distance = float('inf') # start at the highest value possible
max_distance = float('-inf') # start at the lowest value possible

# get the pairwise euclidean distances between all molecules
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        vector_i = data['bag_of_elements'].iloc[i]
        vector_j = data['bag_of_elements'].iloc[j]
        distance = euclidean(vector_i, vector_j) # assume euclidean
        
        # update most similar pair
        if distance < min_distance:
            min_distance = distance
            most_similar_pair = (i, j)
        
        # update most dissimilar pair
        if distance > max_distance:
            max_distance = distance
            most_dissimilar_pair = (i, j)
```

```python
# print it out
print(f"Most similar pair: {most_similar_pair} with a distance: {min_distance}")
print(f"Molecules: {data['smiles'].iloc[most_similar_pair[0]]} and {data['smiles'].iloc[most_similar_pair[1]]}")

print(f"Most dissimilar pair: {most_dissimilar_pair} with a distance: {max_distance}")
print(f"Molecules: {data['smiles'].iloc[most_dissimilar_pair[0]]} and {data['smiles'].iloc[most_dissimilar_pair[1]]}")
```

## **(d)** Plotting distributions.

Plot the distribution of pairwise distances between molecules.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import pdist, squareform

bag_of_elements_array = np.array(data['bag_of_elements'].tolist())

# compute all pairwise Euclidean distances
pairwise_distances = pdist(bag_of_elements_array, metric='euclidean')

# plot the distribution of pairwise distances
plt.figure(figsize=(8, 6))
sns.histplot(pairwise_distances, bins=30, kde=True)
plt.title('distribution of pairwise Euclidean distances between molecules')
plt.xlabel('Euclidean distance')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
```

## **(d)** Standard normalization.

Next calculate and report the average and standard deviation of molecular weights for both the full dataset and the carbonyl-containing subset. Write a new function that takes as input a dictionary and ``standard-normalizes'' the molecular weight. In other words, the entry in the molecular weight will now become

$$\tilde{\text{MW}} = \frac{\text{MW} - \mu_\text{MW}}{\sigma_{\text{MW}}}$$

where $\mu_x$ and $\sigma_x$ indicate the mean and standard deviation of variable $x$.

Plot the distribution of $\tilde{\text{MW}}$. How does this compare to the original distribution?

```python
def standard_normalize_molecular_weight(smiles_dict):
    molar_masses = np.array(list(smiles_dict.values()))
    
    mean_molar_mass = np.mean(molar_masses)
    std_molar_mass = np.std(molar_masses)
    
    # normalize the molecular weights and create a new dictionary
    normalized_smiles_dict = {
        smiles: (molar_mass - mean_molar_mass) / std_molar_mass
        for smiles, molar_mass in smiles_dict.items()
    }
    
    return normalized_smiles_dict
```

```python
# calculate the average and std for full dataset
mean_full_dataset = np.mean(fulldata_molar_masses)
std_full_dataset = np.std(carbonyl_molar_masses)

# calculate avg/std for just carbonyl subset
mean_carbonyl_subset = np.mean(carbonyl_molar_masses)
std_carbonyl_subset = np.std(carbonyl_molar_masses)
```
